```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import datasets
import faker

#iris= datasets.load_iris()
import random
from datetime import datetime, date, timedelta
```

```python
fake=faker.Faker()
```

```python
#setosa dataset
iris = sns.load_dataset("iris")
df=iris.copy()
df["type"]=df.apply(lambda r : random.choice(["Tall","Short","River"]),axis=1)
df["priority"]=df.apply(lambda r : random.choice([1,2,3,4]),axis=1)
df["sample_date"]=df.apply(lambda x: fake.date_between("-6M"),axis=1).astype("datetime64")
df.sort_values("sample_date",ascending=True,inplace=True)
df.info()
```
